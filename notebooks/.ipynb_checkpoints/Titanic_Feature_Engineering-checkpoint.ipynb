```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Ensure df_fe exists before plotting correlation.
if 'df_fe' not in globals():
    df_fe = pd.read_csv('../data/train.csv')

plt.figure(figsize=(14, 10))

# Use numeric columns only to avoid correlation errors with text fields.
numeric_df = df_fe.select_dtypes(include='number')
corr_matrix = numeric_df.corr()

# Create a mask to hide the upper triangle for a cleaner look.
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Plot the heatmap
sns.heatmap(corr_matrix, mask=mask, annot=False, cmap='coolwarm',
            vmax=1, vmin=-1, center=0, square=True, linewidths=.5)

plt.title('Correlation Heatmap of Engineered Features', fontsize=16)
plt.tight_layout()
plt.show()
```
